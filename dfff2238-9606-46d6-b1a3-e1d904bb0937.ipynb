For machine learning, `OneHotEncoder` is much better than `get_dummies`. `get_dummies` increases data size as it converts categorical features into multiple binary columns, which makes the training process more complex and time-consuming. So we should not use it with the tree-based models, since it significantly increases the number of features and computational cost. Instead, we should use  `OneHotEncoder()` from sklearn. What is more, `OneHotEncoder()` handles situations where test subset has features that were not available during training. ## Testing models and their accuracy
To train the models, I will be splitting the data. There will be a training set, a validation set, and a test set. To achieve a 3:1:1 ratio of the sets, we need 60%, 20%, and 20%. I will first split off 20% of the data for the test set. With the remaining 80%, I will split off 25%, because 25% of 80% is 20%. That results with 60-20-20.

```python
#splitting the data into sets
#the first split will be to split off the test set, leaving the trainging and validation sets together
#random state of 616 will be used for this project

data_train_valid, data_test = train_test_split(data_ohe, test_size=0.20, random_state=616) 
data_train, data_valid = train_test_split(data_train_valid, test_size=0.25, random_state=616) # 25% of 80% = 20%
```

### Decision Tree model
<span style='color:blue'>__Here I will be training a decision tree model to predict if a client will leave Beta Bank__</span>

```python
best_depth_dt = 0 
best_auc_roc_dt = 0
best_f1_dt = 0 #establishing variables value of 0 for the for loop to compare to

for depth in range(1, 11):
    model_dt = DecisionTreeClassifier(random_state=616, max_depth=depth) #dt standing for decision tree
    model_dt.fit(features_train, target_train)
    dt_predictions_valid = model_dt.predict(features_valid)
    auc_roc_dt = roc_auc_score(target_valid, dt_predictions_valid)
    f1_dt = f1_score(target_valid, dt_predictions_valid)
    
    if f1_dt > best_f1_dt:
        best_f1_dt = f1_dt
        best_depth_dt = depth
        best_auc_roc_dt = auc_roc_dt
        
print(f"Best F1 Score: {best_f1_dt} Best AUC-ROC Score: {best_auc_roc_dt} Best Depth: {best_depth_dt}")        
```

<span style='color:blue'> __The best decision tree model has an F1 Score of about 0.592 by only needing a depth of 7. That already surpasses our goal score! This gives us a AUC-ROC score of about 0.722__ </span>

### Random Forest Model
<span style='color:green'> __Here I will be training a random forest model to predict if a client will leave Beta Bank__ </span>

```python
best_depth_rf = 0
best_est = 0
best_auc_roc_rf = 0
best_f1_rf = 0

for est in range(10, 51, 10):
    for depth in range(1, 11):
        model_rf = RandomForestClassifier(random_state=616, n_estimators=est, max_depth=depth)
        model_rf.fit(features_train, target_train)
        rf_predicted_valid = model_rf.predict(features_valid)
        auc_roc_rf = roc_auc_score(target_valid, rf_predicted_valid)
        f1_rf = f1_score(target_valid, rf_predicted_valid)
        
        if f1_rf > best_f1_rf :
            best_f1_rf = f1_rf
            best_est = est
            best_depth_rf = depth
            best_auc_roc_dt = auc_roc_rf
print(f"Best F1 Score: {best_f1_rf} Best AUC_ROC Score: {best_auc_roc_dt} Best Number of Estimators: {best_est} Best Depth: {best_depth_rf}")
```

```python
#defining a function to upsample

def upsample(feats, targ, repeat):
    feats_zeros = feats[targ == 0] #split training sample into positive and negative observations
    feats_ones = feats[targ == 1]
    targ_zeros = targ[targ == 0]
    targ_ones = targ[targ == 1]
    
    feats_upsampled = pd.concat([feats_zeros] + [feats_ones] * repeat) 
    targ_upsampled = pd.concat([targ_zeros] + [targ_ones] * repeat) #putting sample back together with multiplied positive observations
    
    feats_upsampled, targ_upsampled = shuffle(feats_upsampled, targ_upsampled, random_state=616) #shuffling the order
    
    return feats_upsampled, targ_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 10)


#because the dataset is new, i will apply a loop to test depths for the decision tree again to see if the best depth is different
best_depth_dt_newest = 0 
best_auc_roc_dt_newest = 0
best_f1_dt_newest = 0 #establishing variables value of 0 for the for loop to compare to

for depth in range(1, 11):
    model_dt_newest = DecisionTreeClassifier(random_state=616, max_depth=depth) 
    model_dt_newest.fit(features_upsampled, target_upsampled)
    dt_predictions_valid_newest = model_dt_newest.predict(features_valid)
    auc_roc_dt_newest = roc_auc_score(target_valid, dt_predictions_valid_newest)
    f1_dt_newest = f1_score(target_valid, dt_predictions_valid_newest)
    
    if f1_dt_newest > best_f1_dt_newest:
        best_f1_dt_newest = f1_dt_newest
        best_depth_dt_newest = depth
        best_auc_roc_dt_newest = auc_roc_dt_newest
        
print(f"Best F1 Score: {best_f1_dt_newest} Best AUC-ROC Score: {best_auc_roc_dt_newest} Best Depth: {best_depth_dt_newest}") 
```

## Downsampling
Next I will try down sampling in attempt to achieve an F1 score of at least 0.59. I will use the new datasets in a decision tree model and a random forest model.

```python
def downsample(feats, targ, fraction): #defined function to downsample data
    feats_zeros = feats[targ == 0] #splitting datasets
    feats_ones = feats[targ == 1]
    targ_zeros = targ[targ == 0]
    targ_ones = targ[targ == 1]
    
    feats_downsampled = pd.concat([feats_zeros.sample(frac=fraction, random_state=616)] + [feats_ones]) #putting datasets back together with sample and frac
    targ_downsampled = pd.concat([targ_zeros.sample(frac=fraction, random_state=616)] + [targ_ones])
    
    feats_downsampled, targ_downsampled = shuffle(feats_downsampled, targ_downsampled, random_state=616) #datasets did the cupid shuffle
    
    return feats_downsampled, targ_downsampled

feats_downsampled, targ_downsampled = downsample(features_train, target_train, 0.256)

#fraction of chosen because the imbalance ratios are 0.7963 (zeros) and 0.2037 (ones)
#I want the proportion of zeros to ones to be approximately equal. So 0.2037/0.7963 = (about) 0.256 

best_depth_dt_super_new = 0 
best_auc_roc_dt_super_new = 0
best_f1_dt_super_new = 0

for depth in range(1, 11):
    model_dt_super_new = DecisionTreeClassifier(random_state=616, max_depth=depth) 
    model_dt_super_new.fit(feats_downsampled, targ_downsampled)
    dt_predictions_valid_super_new = model_dt_super_new.predict(features_valid)
    auc_roc_dt_super_new = roc_auc_score(target_valid, dt_predictions_valid_super_new)
    f1_dt_super_new = f1_score(target_valid, dt_predictions_valid_super_new)
    
    if f1_dt_super_new > best_f1_dt_super_new:
        best_f1_dt_super_new = f1_dt_super_new
        best_depth_dt_super_new = depth
        best_auc_roc_dt_super_new = auc_roc_dt_super_new
        
print(f"Best F1 Score: {best_f1_dt_super_new} Best AUC-ROC Score: {best_auc_roc_dt_super_new} Best Depth: {best_depth_dt_super_new}") 
```

Incredible!! A score of 0.596! Even though it's only more by a few thousandths, it's still a success!! A decision tree with depth 7 on a downsampled dataset.
